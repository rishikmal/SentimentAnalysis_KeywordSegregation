```python
## Sentiment Analysis ALgos Import and classifier Loading
from flair.models import TextClassifier
from flair.data import Sentence

classifier = TextClassifier.load('en-sentiment')

### Text Blob
from textblob import TextBlob
```

```python
## Function for prediction with Flair Model
def flairPred(text):
    if text == "":
        return "Null"
    text = Sentence(text)
    classifier.predict(text)
    value = text.labels[0].to_dict()['value'] 
    return value
```

```python
## Function for prediction with TextBlob Model
def txtBlPred(text):
    if text == "":
        return "Null"
    score = TextBlob(text)
    #print(score.sentiment)
    final_score = score.polarity
    if final_score < 0:
        return "NEGATIVE"
    elif final_score == 0:
        return "NEUTRAL"
    else:
        return "POSITIVE"
```

```python
# import library 
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
import string

# Creating stopwords list which can be removed
stopWordList = stopwords.words('english')
stopWordList += list(string.punctuation)
stopWordList += ["''", '""', '...', '``', 'â€™', 'â€œ', 'â€™', 'â€', 'â€˜', 'â€˜',"'", 'Â©',
'said',"'s", "also",'one',"n't",'com', '-', 'â€“', 'â€”', '_','test', 'testing']
```

```python
## Function for generating Tokens and removing Stop Words
def genToken(data):
    tokens = nltk.word_tokenize(data)
    withoutStopwords = [token.lower() for token in tokens if token.lower() not in stopWordList]
    return withoutStopwords
```

```python
## Using Porter Stemmer for Stemming each Words
from nltk.stem import PorterStemmer
pst = PorterStemmer()# create a function stemming() and loop through each word in a review
def stemm(data):
    stemList = [] #stemmed_review=[]
    for w in data:
        stemList.append(pst.stem(w))
    return stemList
```

```python
## Function for Flatenning the DF as the data recieved is not in appropriate Format
import itertools
def flattenL(df):
    #print(df)
    posList = []
    bool_string = pd.notnull(df)
    posList = df[bool_string]
    posList = list(itertools.chain.from_iterable(posList))
    #print(posList)
    return posList
```

```python
## Function for Segregatting DFs based on their Sentiments
def segregateDF(df):
    dfPositive = df[df['TextBlob'] == 'POSITIVE']
    dfNegative = df[df['TextBlob'] == 'NEGATIVE']
    dfNeutral = df[df['TextBlob'] == 'NEUTRAL']
    return dfPositive, dfNegative, dfNeutral
```

### Main Function for applying Tokenization, Stemming and finally creating final Keyword Lists based on Sentiments

```python
## Main Function for applying Tokenization, Stemming and finally creating final Keyword Lists based on Sentiments
import json
finalDF = pd.DataFrame()
def segregateKeys(df):
    jsonDict = {}
    dfFinalWords = pd.DataFrame()
    df['tokWords'] = df['comment'].apply(lambda x: genToken(x))
    df['stemWords'] = df['tokWords'].apply(lambda x: stemm(x))
    dfPos, dfNeg, dfNeu = segregateDF(df)
    
    posL = flattenL(dfPos['stemWords'])
    neuL = flattenL(dfNeu['stemWords'])
    negL = flattenL(dfNeg['stemWords'])
    dfPos = pd.DataFrame({'POSITIVE':posL})
    dfNeu = pd.DataFrame({'NEUTRAL':neuL})
    dfNeg = pd.DataFrame({'NEGATIVE':negL})
    
    finalDFKeys = pd.concat([dfPos, dfNeu, dfNeg])
    #print(type(finalDFKeys['POSITIVE'].value_counts(dropna=True)))
    vcPos = finalDFKeys['POSITIVE'].value_counts(dropna=True)
    vcNeg = finalDFKeys['NEGATIVE'].value_counts(dropna=True)
    vcNue = finalDFKeys['NEUTRAL'].value_counts(dropna=True)
    #vcNue
    
    finalDF["POSITIVE"] = vcPos#.to_json()
    finalDF['NEGATIVE'] = vcNeg#to_json()
    finalDF['NEUTRAL'] = vcNue#to_json()
        
    return vcPos, vcNeg, vcNue

```
